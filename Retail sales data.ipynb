```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Retail_sales_Worked_data.csv")
```

# Data Visualization

```python
# Bar Plot showing Total Amount per Product Category
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Product_Category', y='Total_Amount', data=df)
plt.title('Total Amount Per Product Category')  # Identify revenue drivers
plt.xlabel('Product Category')
plt.ylabel('Total Amount')
plt.xticks(rotation=45)
plt.show()
```

*This bar plot helps identify which product categories generate the most revenue, allowing for informed decisions on inventory management and marketing strategies.*

```python
# Pie Chart showing Purchase Segmentation distribution
purchase_segmentation_counts = df['Purchase_Segmentation'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(purchase_segmentation_counts, labels=purchase_segmentation_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Purchase Segmentation distribution')
plt.show()
```

*The pie chart visualizes the distribution of purchase segments, which aids in understanding customer demographics and tailoring marketing efforts.*

```python
# Box Plot of Total Amount by Age Group
plt.figure(figsize=(10, 6))
sns.boxplot(x='Age_Group', y='Total_Amount', data=df)
plt.title('Total Amount by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Total Amount')
plt.xticks(rotation=45)
plt.show()
```

*The box plot compares sales across different age groups, identifying which demographics contribute most to total sales.*

```python
# Violin Plot of Quantity distribution by Gender
plt.figure(figsize=(10, 6))
sns.violinplot(x='Gender', y='Quantity', data=df)
plt.title('Quantity Distribution by Gender')  # Inform targeted marketing
plt.xlabel('Gender')
plt.ylabel('Quantity')
plt.show()
```

*The violin plot illustrates how quantity sold varies by gender, informing targeted marketing strategies.*

```python
# Bar Chart of Total Amount trend by Quarter
plt.figure(figsize=(10, 6))
total_amount_by_quarter = df.groupby('Quarter')['Total_Amount'].sum().reset_index()
sns.barplot(x='Quarter', y='Total_Amount', data=total_amount_by_quarter)
plt.title('Total Amount Trend by Quarter')  # Assess quarterly performance
plt.xlabel('Quarter')
plt.ylabel('Total Amount')
plt.show()
```

*The Bar Chart summarizes sales trends by quarter, which helps in assessing quarterly performance and planning for financial reviews.*

```python
# Bar Chart showing Total Amount over Time
df['Date'] = pd.to_datetime(df['Date'])
df['Month_Year'] = df['Date'].dt.to_period('M')

plt.figure(figsize=(12, 6))
total_amount_over_time = df.groupby('Month_Year')['Total_Amount'].sum().reset_index()
sns.barplot(x='Month_Year', y='Total_Amount', data=total_amount_over_time)
plt.title('Total Amount over Time')  # Analyze trends for forecasting
plt.xlabel('Month_Year')
plt.ylabel('Total Amount')
plt.xticks(rotation=45)
plt.show()
```

*The Bar Chart illustrates trends in sales over time, providing insights into seasonal patterns and helping to forecast future sales.*
